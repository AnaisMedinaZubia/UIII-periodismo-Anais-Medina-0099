#### Diccionario Periodista
## Periodista: Medio de comunicación, fuente informativa, artículos publicados, ética profesional, acreditación de prensa y región.

```python
# ==============================
# Diccionario de Periodistas
# ==============================

# Diccionario vacío
Diccionario_periodista = []

# ==============================
# Función para mostrar tabla
# ==============================
def Listar_periodista():

    print("\n==============================================================")
    print("                 LISTA DE PERIODISTAS")
    print("==============================================================")
    print(f"{'ID':<5} {'NOMBRE':<20} {'MEDIO':<20} {'EDAD':<10}")
    print("--------------------------------------------------------------")

    for periodista in Diccionario_periodista:
        for key, value in periodista.items():
            pass

        print(f"{periodista['id']:<5} "
              f"{periodista['nombre']:<20} "
              f"{periodista['medio']:<20} "
              f"{periodista['edad']:<10}")

    print("==============================================================\n")


# ==============================
# Función agregar
# ==============================
def agregar_periodista():

    id_periodista = input("Ingrese ID: ")
    nombre = input("Ingrese nombre: ")
    medio = input("Ingrese medio de comunicación: ")
    edad = input("Ingrese edad: ")

    periodista = {
        "id": id_periodista,
        "nombre": nombre,
        "medio": medio,
        "edad": edad
    }

    Diccionario_periodista.append(periodista)

    print("\nPeriodista agregado correctamente.")
    Listar_periodista()


# ==============================
# Función buscar
# ==============================
def buscar_periodista():

    buscar = input("Ingrese el nombre del periodista a buscar: ")

    print("\n==============================================================")
    print("                 PERIODISTA ENCONTRADO")
    print("==============================================================")
    print(f"{'ID':<5} {'NOMBRE':<20} {'MEDIO':<20} {'EDAD':<10}")
    print("--------------------------------------------------------------")

    for periodista in Diccionario_periodista:

        if periodista["nombre"] == buscar:

            for key, value in periodista.items():
                pass

            print(f"{periodista['id']:<5} "
                  f"{periodista['nombre']:<20} "
                  f"{periodista['medio']:<20} "
                  f"{periodista['edad']:<10}")

    print("==============================================================\n")


# ==============================
# Función borrar
# ==============================
def borrar_periodista():

    borrar = input("Ingrese nombre del periodista a borrar: ")

    for periodista in Diccionario_periodista:

        if periodista["nombre"] == borrar:
            Diccionario_periodista.remove(periodista)
            print("\nPeriodista eliminado correctamente.")

    Listar_periodista()


# ==============================
# Función actualizar
# ==============================
def actualizar_periodista():

    actualizar = input("Ingrese nombre del periodista a actualizar: ")

    for periodista in Diccionario_periodista:

        if periodista["nombre"] == actualizar:

            periodista["nombre"] = input("Nuevo nombre: ")
            periodista["medio"] = input("Nuevo medio: ")
            periodista["edad"] = input("Nueva edad: ")

            print("\nDatos actualizados correctamente.")

    Listar_periodista()

# ==============================
# Menú interactivo
# ==============================

while True:

    print("=============== MENÚ ===============")
    print("1. Agregar periodista")
    print("2. Listar periodistas")
    print("3. Buscar periodista")
    print("4. Borrar periodista")
    print("5. Actualizar periodista")
    print("6. Salir")
    print("====================================")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        agregar_periodista()

    elif opcion == "2":
        Listar_periodista()

    elif opcion == "3":
        buscar_periodista()

    elif opcion == "4":
        borrar_periodista()

    elif opcion == "5":
        actualizar_periodista()

    elif opcion == "6":
        print("\nPrograma finalizado.")
        print("Programa realizado por Anais Medina NL = 36")
        break

    else:
        print("\nOpción incorrecta.\n")
```
